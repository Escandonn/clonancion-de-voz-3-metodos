ğŸ§¹ 4. Normalizar audios (MUY IMPORTANTE)


```
Qwen funciona mejor a 16kHz mono.
```



````python
def limpiar_audio(input_path, output_path):
    audio, sr = librosa.load(input_path, sr=16000, mono=True)
    sf.write(output_path, audio, 16000)

limpiar_audio("mi_voz.wav", "mi_voz_clean.wav")
limpiar_audio("audio_estilo.wav", "audio_estilo_clean.wav")
````
