Importar librerías y cargar modelo (modo correcto para Colab)

```python
import torch
from qwen_tts import Qwen3TTSModel

model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-1.7B-Base",
    device_map="cuda:0",
    dtype=torch.bfloat16,
    attn_implementation="sdpa",  # importante en Colab
)
```

cargar texto

```python
ref_audio = "(nombre_de_tu_nuevo_archivo.wav)"
ref_text = """
(Aquí va la transcripción exacta de tu nuevo archivo de audio)
"""
```

Generar audio con tu voz clonada

```python
import soundfile as sf
from IPython.display import Audio, display
from google.colab import files

# Instalar pydub para la conversión a MP3
!pip install -q pydub
from pydub import AudioSegment

text = """
(Por favor, ingresa aquí el nuevo texto que quieres que el clon diga.)
"""

wavs, sr = model.generate_voice_clone(
    text=text,
    language="Spanish",
    ref_audio=ref_audio,
    ref_text=ref_text,

    # 🎭 CONTROL DE EMOCIÓN (aquí está el secreto)
    instruct = """
Energetic modern commercial voice for a tech startup.

Young male voice, charismatic, friendly, innovative tone,
fast but clear pacing, digital entrepreneur vibe,
motivational and fresh style, persuasive and inspiring.

Sounds like a cutting-edge web solutions brand.
""",

    # 🔥 Parámetros que afectan la emoción y actuación
    do_sample=True,
    temperature=0.7, # Aumentado ligeramente para mayor variabilidad emocional
    top_p=0.85, # Reducido ligeramente para enfocar en tokens más probables pero aún expresivos
    repetition_penalty=1.2, # Aumentado para reducir la monotonía
    max_new_tokens=1024,
)

# 🔊 Escuchar en Colab
display(Audio(wavs[0], rate=sr))

# 💾 Guardar como WAV
wav_filename = "voz_clonada_emocion.wav"
sf.write(wav_filename, wavs[0], sr)

# Convertir WAV a MP3
mp3_filename = "voz_clonada_emocion.mp3"
audio = AudioSegment.from_wav(wav_filename)
audio.export(mp3_filename, format="mp3")

# ⬇️ Descargar el archivo MP3
files.download(mp3_filename)
```
