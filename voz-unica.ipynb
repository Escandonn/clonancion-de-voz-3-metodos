```python
import torch
import soundfile as sf
from qwen_tts import Qwen3TTSModel
from google.colab import files

```

üé§ OPCI√ìN 1 ‚Äî Custom Voice (voces predefinidas tipo Ryan, Vivian, etc.)
Cargar modelo

```python
model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-1.7B-VoiceDesign",
    device_map="cuda:0",
    dtype=torch.bfloat16,
    attn_implementation="flash_attention_2",
)

```

Generar audio

```python
from IPython.display import Audio, display
import soundfile as sf
from google.colab import files

# üìù Texto protegido y formateado con pausas naturales
text = """
Baj√° la voz‚Ä¶ mir√° que aqu√≠, hasta las paredes, oyen.

No hag√°s ruido.
Camin√° suave.
Respir√° lento.

Que nadie note que estamos aqu√≠.
Que nadie sospeche‚Ä¶ absolutamente nada.

Tranquilo.
Paso a paso.
Sin af√°n.

Porque en este juego, el que se apura‚Ä¶ pierde.

Y nosotros,
no vinimos a perder.
"""

# üéôÔ∏è Instrucci√≥n optimizada (mezcla espa√±ol + ingl√©s = mejor resultado)
instruct = (
    "Colombian male voice with strong Paisa accent from Medellin, "
    "very deep and masculine tone, intimidating mafia boss style, "
    "slow confident rhythm, warm but dominant presence."
)

# üöÄ Generaci√≥n PRO
wavs, sr = model.generate_voice_design(
    text=text,
    language="Spanish",
    instruct=instruct,

    # üîí Par√°metros cr√≠ticos de fidelidad
    do_sample=False,
    temperature=0.2,
    repetition_penalty=1.25,
    top_p=0.8,
    max_new_tokens=1024,
)

# üîä Escuchar directamente en Colab
display(Audio(wavs[0], rate=sr))

# üíæ Guardar archivo
filename = "voice_design_pro.wav"
sf.write(filename, wavs[0], sr)

# ‚¨áÔ∏è Descargar
files.download(filename)


```
